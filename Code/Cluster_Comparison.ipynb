```python
import numpy as np
import pandas as pd
import csv
```

```python
data = pd.read_csv(r"C:\Users\colef\OneDrive - University of Miami\Documents\College\Research\Wuchty Lab\Data\HI-union.tsv",sep='\t')
data.drop(data[(data['Protein1'] == data['Protein2'])].index,inplace=True)
data.reset_index(drop=True,inplace=True)
protein_set = set([*data.Protein1,*data.Protein2])
protein_list = list(protein_set)
```

```python
TADdata = pd.read_csv(r"C:\Users\colef\OneDrive - University of Miami\Documents\College\Research\Wuchty Lab\Data\hg38_GM12878_Lieberman_2009-raw_TADs_with_gene_ids.tsv",sep='\t')
TADdata.drop(['chr','start','end'],axis=1,inplace=True)

def process(gene):
    gene = str(gene)
    gene = gene.split('.', 1)[0]
    if gene in protein_list:
        return gene
    else:
        return np.nan

TADdata = TADdata.map(process)
TADdata = TADdata.to_numpy()
```

```python
TADgenes = []
for row in TADdata:
    temp = row[~pd.isna(row)]
    TADgenes.append(temp)
TADgenes
```

```python
file = open(r"C:\Users\colef\OneDrive - University of Miami\Documents\College\Research\Wuchty Lab\Data\HI-union_compressedV2.csv")
TADproteins = list(csv.reader(file,delimiter=','))
file.close()

TADproteins = [np.array(protein[0].split('+'),dtype=object) for protein in TADproteins]
```

```python
running_sum = 0
for supernode in TADproteins:
    running_sum += len(supernode)
print(1 - (len(TADproteins) / running_sum))

running_sum = 0
for supernode in TADgenes:
    running_sum += len(supernode)
print(1 - (len(TADgenes) / running_sum))
```
